```python
import scipy
import numpy as np
import matplotlib.pyplot as plt
```

```python
def brownian_kernel(x1, x2, amp = 1):
    K = np.zeros((len(x1), len(x2)))
    for i in range(len(x1)):
        for j in range(len(x2)):
            K[i,j] = min(x1[i], x2[j])
    return amp * K

x = np.linspace(0, 1, 100)
K = brownian_kernel(x, x)
f = scipy.stats.multivariate_normal.rvs(mean = np.zeros(len(x)), cov = K, size = 5)

for func in f:
    plt.plot(x, func)
plt.show()
```

```python
def squared_exponential_kernel(x1, x2, length_scale = 1, amp = 1):
    K = amp**2 * np.exp(-(x1[:, np.newaxis] - x2[np.newaxis, :])**2 / 2 / length_scale**2)
    return K

x = np.linspace(0, 1, 100)
K = squared_exponential_kernel(x, x, length_scale=0.1)
f = scipy.stats.multivariate_normal.rvs(mean = np.zeros(len(x)), cov = K, size = 5)

for func in f:
    plt.plot(x, func)
plt.show()
```
